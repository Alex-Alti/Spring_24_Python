```python
# Assign gap
gap = 4
# A
a1 = n[1]*5
a2 = n[1]+n[0]*3+n[1]
a3 = n[1]*5
a4 = n[1]+n[0]*3+n[1]
a5 = n[1]+n[0]*3+n[1]

# L
l1 = n[0]*3+n[2]
l2 = n[0]*3+n[2]
l3 = n[0]*3+n[2]
l4 = n[0]*3+n[2]
l5 = n[2]*5

# E
e1 = n[0]*7+n[3]*5
e2 = n[0]*7+n[3]
e3 = n[0]*7+n[3]*5
e4 = n[0]*7+n[3]
e5 = n[3]*5

# X
x1 = n[0]*3+n[4]+n[0]*3+n[4]
x2 = n[0]*8+n[4]+n[0]+n[4]
x3 = n[0]*5+n[4]
x4 = n[0]*8+n[4]+n[0]+n[4]
x5 = n[4]+n[0]*3+n[4]

# Create lines
line1 = [a1,l1,e1,x1]
line2 = [a2,l2,e2,x2]
line3 = [a3,l3,e3,x3]
line4 = [a4,l4,e4,x4]
line5 = [a5,l5,e5,x5]

s1 = (n[0]).join(line1)
s2 = (n[0]).join(line2)
s3 = (n[0]).join(line3)
s4 = (n[0]).join(line4)
s5 = (n[0]*gap).join(line5)
s = [s1,s2,s3,s4,s5]

print('\n'.join(s))
```
